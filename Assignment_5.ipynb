```python
import matplotlib.pyplot as plt
import numpy as np

# Defining the matrix
matrix = [
    [None, 72, 41, 42, 43, 44, 45, 46, None],
    [70, 71, 40, 17, 18, 19, 20, 47, 48],
    [69, 38, 39, 16, 1, 2, 21, 22, 49],
    [68, 37, 14, 15, None, 3, 4, 23, 50],
    [67, 36, 13, None, None, None, 5, 24, 51],
    [66, 35, 12, 11, None, 7, 6, 25, 52],
    [65, 34, 33, 10, 9, 8, 27, 26, 53],
    [64, 63, 32, 31, 30, 29, 28, 55, 54],
    [None, 62, 61, 60, 59, 58, 57, 56, None]
]

# Initialize lists to store coordinates
odd_coords, even_coords = [], []

# Assigning coordinates and separating odd and even numbers
for n, row in enumerate(matrix):
    for m, value in enumerate(row):
        if value is not None:
            if value % 2 == 0:  # Even
                even_coords.append((m - 4, 4 - n))
            else:  # Odd
                odd_coords.append((m - 4, 4 - n))

# Plotting odd numbers
plt.figure(figsize=(8, 8))
plt.scatter(*zip(*odd_coords), color='red', label='Odd Numbers')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.title('Odd Numbers in the Matrix')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.show()

# Plotting even numbers
plt.figure(figsize=(8, 8))
plt.scatter(*zip(*even_coords), color='blue', label='Even Numbers')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.title('Even Numbers in the Matrix')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.show()

# Plotting both odd and even numbers together
plt.figure(figsize=(8, 8))
plt.scatter(*zip(*odd_coords), color='red', label='Odd Numbers')
plt.scatter(*zip(*even_coords), color='blue', label='Even Numbers')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.title('Odd and Even Numbers in the Matrix Combined')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.show()
```

```python
from sympy import isprime

# Initialize list to store coordinates of primes
prime_coords = []

# Assigning coordinates for prime numbers 
for n, row in enumerate(matrix):
    for m, value in enumerate(row):
        if value is not None and isprime(value):
            # Calculating coordinates based on the user's correction
            prime_coords.append((m - 4, 4 - n))
            
# Extracting x and y coordinates for plotting
x_prime, y_prime = zip(*prime_coords)

# Plotting the corrected prime coordinates
plt.figure(figsize=(8, 8))
plt.scatter(x_prime, y_prime, color='blue', label='Primes')
plt.title('Prime Number Coordinates (Corrected)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import math

# Initialize list to store coordinates of square numbers
square_coords = []

# Assigning coordinates for square numbers using the corrected matrix and method
for n, row in enumerate(matrix):
    for m, value in enumerate(row):
        if value is not None and math.isqrt(value)**2 == value:
            # Calculating coordinates for square numbers
            square_coords.append((m - 4, 4 - n))

# Extracting x and y coordinates for plotting square numbers
x_square, y_square = zip(*square_coords)

# Plotting the corrected square number coordinates
plt.figure(figsize=(8, 8))
plt.scatter(x_square, y_square, color='red', label='Squares')
plt.title('Square Number Coordinates')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
def S(n):
    return 1 + sum([4*(2*i + 2) for i in range(1, n)])

def E(n):
    return sum([4*(2*i + 2) for i in range(1, n+1)])
```

```python
def R_n_1(n):
    return [E(n)] + list(range(S(n), S(n) + 2*n))

def R_n_2n_plus_3(n):
    start = S(n) + 4*n + 3
    end = S(n) + 6*n + 4  
    return list(range(start, end))

# calculating for n=4
R_4_1 = R_n_1(4)
R_4_11 = R_n_2n_plus_3(4)

# Print statements for the values
print(f"S(4): {S(4)}")
print(f"E(4): {E(4)}")
print(f"R_4(1): {R_4_1}")
print(f"R_4(11): {R_4_11}")
```
